## First Import all the necessary library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Then import the data 

```python
df = pd.read_csv(r'Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df
```

## Check the shape of data 

```python
df.shape
```

```python
df.info()  # to get the info of data 
```

## here we know the Status and Unnamed column is not needed or not necessary so its deleted

```python
df.drop(['Status','unnamed1'], axis = 1 , inplace = True)
```

## Then check the null value 

```python
df.isnull().sum()
```

## Now we remove null value like in amount column 12 null value 

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()  # Check its remove ?
```

## Now we See the Amount coloumn data type is float now we can convert into int

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

## Now we see the summary of data 

```python
df[['Age','Orders','Amount']].describe()  # only Necessary columns
```

## Lets Start with basic EDA 

## First we take countplot of Gender

```python
v = sns.countplot(x = 'Gender', data = df);

for bar in v.containers:
    v.bar_label(bar)
```

## The above graph shows the Female Customers is more purchasing 

## Lets see the purchasing amount based on gender first we a groupby 

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen
```

## Lets create Graph

```python
v = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen);
```

From above graph we can see the most of the buyers are females and even the purchasing power is greater then men

## Create a graph for AGE group on gender

```python
val = sns.countplot(x = 'Age Group',data = df, hue = 'Gender')
for b in val.containers:
    val.bar_label(b)
```

## Here we the purchasing amount based on age group 

```python
sales_age = df.groupby(['Age Group', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_age
```

```python
sns.barplot(x = 'Age Group', y = 'Amount',hue = 'Gender',data = sales_age);
```

From above graph we can see that most of the buyers are age group of 26-35 that are FEMALE

## Now we see total number of order from states

```python
sales_state = df.groupby('State',as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
```

```python
sales_state
```

```python
sns.set(rc={'figure.figsize': (15,6)})
sns.barplot(x='State', y='Orders', data=sales_state)
plt.xticks(rotation=45)
plt.show()
```

## now we see the amount purchasing of state

```python
sale_state_am = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sale_state_am
```

```python
sns.set(rc={'figure.figsize': (15,6)})
sns.barplot(x='State', y='Amount', data=sale_state_am)
plt.xticks(rotation=45)
plt.show()
```

From above graph we can see that most of the orders& total sales/amounnt from UttarPradesh,Maharashtra and Karnataka 

## Maritial Status of who are shopping

```python
sns.set(rc={'figure.figsize': (2,5)})
m = sns.countplot(x = 'Marital_Status', data = df)

for bar in m.containers:
    m.bar_label(bar)
```

## Now we see Marital status who purchace a lot based on gender 


```python
ma = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
ma
```

```python
sns.set(rc={'figure.figsize': (5,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', hue = 'Gender', data = ma);
```

From above graph we see most of the buyers are married women and its purchasing power also high

## Now see which occupation sector order most 

```python
sns.set(rc={'figure.figsize': (15,5)})
oc = sns.countplot(x = 'Occupation', data = df)

for o in oc.containers:
    oc.bar_label(o)

plt.xticks(rotation=45)
plt.show()
```

```python
occ = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
occ
```

```python
sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(x = 'Occupation',y='Amount', data = occ)
plt.xticks(rotation=45)
plt.show()
```

From above graph we can see that most of the buyers are from IT Sector, HealthCare and Aviation

## Lets see the product category who sale more

```python
sns.set(rc={'figure.figsize': (20,5)})
p = sns.countplot(x = 'Product_Category', data = df)

for c in p.containers:
    p.bar_label(c)

plt.xticks(rotation=45)
plt.show()
```

```python
pc = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
pc
```

```python
sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(x = 'Product_Category',y='Amount', data = pc)
plt.xticks(rotation=45)
plt.show()
```

From above Graph we can see that most of the sold products are Food , Clothing and Electronic & Gadgets

## Lets see top 10 products who are order most by Product - ID


```python
pid = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
pid
```

```python
sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(x = 'Product_ID',y='Orders', data = pid)
plt.xticks(rotation=45)
plt.show()
```
